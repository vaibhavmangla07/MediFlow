# Feature Summary

```python
feature_summary = pd.DataFrame({
    "Data Type": df.dtypes,
    "Missing Values": df.isnull().sum(),
    "Unique Values": df.nunique(),
    "Duplicated Rows": df.duplicated().sum(),
    "Memory Usage (MB)": round(df.memory_usage(deep=True)/1024**2, 2), 
})

feature_summary
```

## Data Cleaning

```python
# Drop Unnecessary Columns
df.drop(columns=["Call_ID"], inplace=True)
```

```python
# Handle Missing Values
df["Chronic_Condition"] = df["Chronic_Condition"].fillna("No Chronic Condition")
df.isnull().sum()
```

```python
print("Shape of the dataset after handling missing values:", df.shape)
df.head()
```

# Save Clean Dataset

```python
df.to_csv("../dataset/cleaned_data.csv", index=False)
print("Cleaned dataset saved to 'cleaned_data.csv'")
```
